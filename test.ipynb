```python
# webスクレイピングに最低限必要な対ブラリのインポート
import requests
from bs4 import BeautifulSoup
```

# HTTPリクエスト
## 基本のリクエスト

```python
# アクセスしたいwebサイトのURLを設定
url = 'https://www.musashino-u.ac.jp/'
# webサーバーにHTTPリクエストを送信
# レスポンスを変数に格納しておく
res = requests.get(url)
```

```python
res.encoding = res.apparent_encoding  # レスポンスのエンコーディングを設定
print(f"レスポンス：{res}") # レスポンスオブジェクトを表示
print(f"レスポンスの型：{type(res)}") # レスポンスオブジェクトの型を表示
print(f"ステータスコード：{res.status_code}") # ステータスコードを表示
print(f"ステータスメッセージ：{res.reason}") # ステータスメッセージを表示

print(f"リクエスト：{res.request}") # リクエストオブジェクトを表示
print(f"リクエストの型：{type(res.request)}") # リクエストオブジェクトの型を表示
print(f"リクエストヘッダー：{res.request.headers}") # リクエストヘッダーを表示
print(f"リクエストメソッド：{res.request.method}") # リクエストメソッドを表示
print(f"リクエストURL：{res.request.url}") # リクエストURLを表示
```

```python
# レスポンス
res.encoding = res.apparent_encoding  # レスポンスのエンコーディングを設定
print(f"レスポンス：{res.headers}") # レスポンスヘッダーを表示
print(f"レスポンスボディ：{(res.text)}") # レスポンスボディの型を表示
```

# パラメータ付きの

```python
# url = "https://www.musashino-u.ac.jp/search/html?query=データサイエンス"
url = "https://www.musashino-u.ac.jp/search/html"
params = {
    'query': 'データサイエンス'
}
res = requests.get(url, headers=headers, params=params)
res.encoding = res.apparent_encoding 
print(f"ステータスコード：{res.status_code}") # ステータスコード
print(f"レスポンスボディ：{(res.text)}") # レスポンスボディの型を表示
```

# BeautifulSoupでHTMLソースから任意の情報を取得する


```python
url = "https://www.musashino-u.ac.jp/"
res = requests.get(url, headers=headers)
res.encoding = res.apparent_encoding
print(f"ステータスコード：{res.status_code}") # ステータスコード
```

```python
# HTMLのソースコードを解析
soup = BeautifulSoup(res.text, 'html.parser')
print(f"soupの型：{type(soup)}")  # soupオブジェクトの方を表示
```

```python
# ソースコード全体から<a>タグを抽出
soup_a = soup.a
print(f"<a>タグ：{soup_a}")  # 最初に見つかった<a>タグを表示
print(f"<a>タグの型：{type(soup_a)}")  # <a>タグオブジェクトの方を表示

# ソースコード全体から<a>タグをすべて抽出
soup_all_a = soup.find_all('a')
print(f"すべての<a>タグ：{soup_all_a}")  # すべての<a>タグを表示
```

```python
# ソースコード全体から<dig>タグを抽出
soup_div = soup.div
# print(f"<div>タグ：{soup_div}")  # 最初に見つかった<div>タグを表示
# print(f"<div>タグの型：{type(soup_div)}")  # <div>タグオブジェクトの方を表示

soup_div_div = soup_div.div
print(f"<div>タグの中の<div>タグ：{soup_div_div}")  # 最初に見つかった<div>タグを表示
print(f"<div>タグの中の<div>タグの型：{type(soup_div_div)}")  # <div>タグオブジェクトの方を表示
```

```python
soup_div_div = soup_div.div
print(f"<div>タグの中の<div>タグ：{soup_div_div}")  # 最初に見つかった<div>タグを表示
print(f"<div>タグの中の<div>タグの型：{type(soup_div_div)}")  # <div>タグオブジェクトの方を表示 

    
```

```python
# Tagオブジェクト.stringでタグ内のテキストを取得できる。
print(f"<a>タグのテキスト：{soup_a.string}")  # <a>タグ内のテキストを表示

# コンテンツがない場合はNoneが返される
print(f"<div>タグの中の<div>タグのテキスト：{soup_div_div.string}")  # <div>タグ内のテキストを表示
```

## 指定してたHTMLタブで要素を検索

```python
# 指定したタグが最初に使われている要素のうち、最初に見つかった要素を取得する。
h2 = soup.find('h2')
print(f"<h2>タグ{h2}")  # 最初に見つかった<h2>タグを表示
print(f"<h2>変数の型：{type(h2)}")  # <h2>タグオブジェクトの型を表示
```

```python
h2_all = soup.find_all('h2')
print(f"<h2>タグのリスト:{h2_all}")  # すべての<h2>タグを表示
print(f":{type(h2_all)}")  # <h2>タグオブジェクトのリストの型を表示 
```

```python
a_list = soup.find_all('a')
for a in a_list:
    print(f"{a.string}") # 各<a>タグ内のテキストを表示
# a_list = soup.find_all('a')
# for a in a_list:
# 検索結果の数に制限をかける
a_list_mlimited = soup.find_all('a', limit=5)
for a in a_list_mlimited:
    print(f"{a.string}") # 各<a>タグ内のテキストを表示
    

```

```python
# 複数のタグをリストで指定して検索することも可能
# 例えば、<h2>と<h3>タグの両方を

h2_h3_list = soup.find_all(['h2', 'h3'])
for tag in h2_h3_list:
    print(f"{tag.s}")  # 各<h2>および<h3>タグ内のテキストを表示
```

## 任意の属性で検索

```python
tag_href_list = soup.find_all(attrs={'href': '/'}, limit=5)
for tag in tag_href_list:
    print(f"{tag}")  # href属性が'/'のタグを表示
```

```python
# id属性やclass属性など、任意の属性で検索することも可能
# 例えば、id属性が'main-content'の要素を検索する場合は以下のようにする.
tag_id_list = soup.find_all(attrs={'id': 'top'})
for tag in tag_id_list:
    print(f"{tag}")  # id属性が'top'のタグを表示
tag_class_list = soup.find_all(class_='p-notices', limit=5)
for tag in tag_class_list:
    print(f"{tag}")  # class属性が'p-notices'のタグを表示
```

## CSSセレクトを使った検索

```python
# id属性を指定（#＋　検索id名）
id_tags = soup.select('#top')
for tag in id_tags:
    print(f"{tag}")  # id属性が'top'のタグを表示
## class属性を指定（.+ 検索クラス名）
# class_tags = soup.select('.p-notices')
# for tag in class_tags:
#     print(f"{tag}")  # class属性が'p-notices'のタグを

## HTML階層構造を指定
# 例えば、<div>タグの中にある<a>タグを検索する場合は以下のようにする.                                                                           
div_h2_tags = soup.select('body')
for tag in div_h2_tags:
    print(f"{tag}")  # <div>タグの中にある<h2>タグを表示
```
